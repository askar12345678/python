```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.gridspec as gridspec 
```

```python
def func(x):
    return (np.sin(x/5)*np.exp(x/10) + 5*np.exp(-x/2))
```

```python
a = np.array([[1, 1], [1, 15]])
b = np.array([func(1), func(15)])
u = np.linalg.solve(a, b)
w0 = u[0]
w1 = u[1]

def pol1(x):
    return w0 + w1*x
x = np.arange(1, 16, 0.01)
y1 = func(x)
y2 = pol1(x)
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('X', fontsize=14) 
plt.ylabel('Y', fontsize=14, rotation = 0)
plt.show()
```

Плохо приближает

```python
a = np.array([[1, 1, 1], [1, 8, 64], [1, 15, 225]])
b = np.array([func(1), func(8), func(15)])
u = np.linalg.solve(a, b)

w0 = u[0]
w1 = u[1]
w2 = u[2]
def pol2(x):
    return w0 + w1*x + w2*x*x
x = np.arange(1, 16, 0.01)
y1 = func(x)
y2 = pol2(x)
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('X', fontsize=14) 
plt.ylabel('Y', fontsize=14, rotation = 0)
plt.show()

```

Плохо приближает, но немного получше линейного

```python
a = np.array([[1, 1, 1, 1], [1, 4, 16, 64], [1, 10, 100, 1000], [1, 15, 225, 3375]])
b = np.array([func(1), func(4), func(10), func(15)])
u = np.linalg.solve(a, b)

w0 = u[0]
w1 = u[1]
w2 = u[2]
w3 = u[3]
def pol3(x):
    return w0 + w1*x + w2*x*x + w3*x*x*x
x = np.arange(1, 16, 0.01)
y1 = func(x)
y2 = pol3(x)
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('X', fontsize=14) 
plt.ylabel('Y', fontsize=14, rotation = 0)
plt.show()
print(u)
```

Приближает гораздо лучше чем предыдущие

```python
def tlr(n, x0):
    return ((-1)**(n+1))*(x0**n)/n
```

```python
n = int(input())
x0 = float(input())
arr = np.arange(1, n+1)
arr = tlr(arr, x0)
print(np.sum(arr))
```

```python
X = np.array([[175, 4], [20, 2], [25, 8]])
print(X)

mean1 = np.sum(X[: , 0]) / 3
mean2 = np.sum(X[: , 1]) / 3
mean = np.array([mean1 , mean2])

x = X - mean
M = (x[: , 0]) * (x[: , 1])

cov12 = np.sum(M) / 3
cov_matrix = np.array([[np.var(X[: , 0]), cov12] , [cov12, np.var(X[: , 1])]])
print(cov_matrix)
```
